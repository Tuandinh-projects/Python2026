### Task 1: A list of dictionaries (1p)

You are given a list of raw transaction records, where each record is a dictionary:

```python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]
```

Write a script that performs the following steps:

a) Filter out invalid transactions where the price is zero or the quantity is negative.

b) For each remaining transaction, compute a new field `total` = `price * quantity`.

c) Group the valid transactions by `region` and compute the total revenue per region. Store the result in a dictionary.

d) Print the region with the highest total revenue and its value using an f-string.

````python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]

# a)
print("a)")
for i in transactions:
    if i["price"] == 0 or i["quantity"] < 0:
       print(f"{i}")        

# b)
print()
print("b)")
for i in transactions:
    if i["price"] != 0 and i["quantity"] > 0:
        total = i["price"]*i["quantity"]
        print(f"{i["product"]}, total = {total}")  

# c)
print()
print("c)")
revenue_by_region = {}

for i in transactions:
    region = i["region"]
    if i["price"] != 0 and i["quantity"] > 0:  #filter valid transactions
        total = i["price"] * i["quantity"]
        #add value by region to the dictionary, if not exist -> create new item, if exist -> add to that item
        if region not in revenue_by_region:    
            revenue_by_region[region] = total  
        else:
            revenue_by_region[region] += total
print(revenue_by_region)

# d)
print()
print("d)")
max_region = max(revenue_by_region, key=revenue_by_region.get) #get region with highest revenue
max_revenue = max(revenue_by_region.values()) #get the highest revenue by region
print(f"Region with highest revenue is {max_region} with {max_revenue}")
````

### Task 2: Creating a function (1p)

You are given the following list of daily sales figures (in euros) for a two-week period:

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]
```

Write a function called `summarise_sales` that takes a list of daily sales values and returns a dictionary containing:

`"total"`: the sum of all daily sales

`"average"`: the mean daily sales value, rounded to two decimals

`"best_day"`: the index (starting from 1) of the day with the highest sales

`"worst_day"`: the index (starting from 1) of the day with the lowest sales

`"above_average_count"`: the number of days where sales exceeded the average

Call the function with the list above and print the returned dictionary.

````python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

def summarise_sales(sales):
    total = sum(sales)
    average = round(total/len(sales),2)
    best_day = sales.index(max(sales)) + 1
    worst_day = sales.index(min(sales)) +1
    # create a loop to count the number of days with sales above avg, store in the variable "above_average_count"
    above_average_count = 0
    for x in sales:
        if x > average:
            above_average_count = above_average_count + 1  
    # return the above calculation in the form of a dictionary
    return {
        "total": total,
        "average": average,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": above_average_count
    }

summarise_sales(sales)
````

### Task 4: Class and methods (1p)

Design a class called `Portfolio` that represents an investment portfolio. The constructor is provided for you:

```python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}
```

Each key in `holdings` is a ticker symbol (string) and each value is a dictionary with keys `"shares"` and `"price_per_share"`. Add the following methods to the class:

a) A method `add_asset(ticker, shares, price_per_share)` that adds a new asset or updates the number of shares if the ticker already exists (in which case the price should also be updated).

b) A method `total_value()` that returns the total market value of the portfolio.

c) A method `most_valuable()` that returns the ticker of the asset with the highest total value (shares × price).

d) A `__str__` method that returns a readable summary of the portfolio, including the owner's name, the number of assets, and the total value.

Create an instance, add at least three assets, and demonstrate all methods with printed output.

````python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}
    # a)
    def add_asset(self, ticker, shares, price_per_share):
        if ticker in self.holdings:
            old_price = self.holdings[ticker]["price_per_share"] #calculate old price
            old_shares = self.holdings[ticker]["shares"] 
            total_shares = old_shares + shares # calculate total
            avg_price = (old_shares*old_price + shares*price_per_share)/total_shares #calculate avg share price
            self.holdings[ticker]["shares"] += shares #adding more shares to current holdings
            self.holdings[ticker]["price_per_share"] = avg_price #update new price
        else:
            self.holdings[ticker] = {
                "shares": shares,
                "price_per_share": price_per_share
            }

    # b)
    def total_value(self):
        total_value = 0
        for ticker in self.holdings:
            total_value += self.holdings[ticker]["shares"]*self.holdings[ticker]["price_per_share"]
        return total_value

    # c)
    def most_valuable(self):
        max_value = 0
        max_ticker = None
        for ticker in self.holdings:
            if self.holdings[ticker]["shares"]*self.holdings[ticker]["price_per_share"] > max_value:
                max_value = self.holdings[ticker]["shares"]*self.holdings[ticker]["price_per_share"]
                max_ticker = ticker
        return max_ticker

    # d)
    def __str__(self):
            return (
                f"Portfolio owner: {self.owner}\n"
                f"Number of assets: {len(self.holdings)}\n"
                f"Total value: {self.total_value():.2f}€"
            )
        
    
    
my_invest = Portfolio("Tuan")
my_invest.add_asset("AAPL", 10, 150)
my_invest.add_asset("TSLA", 5, 700)
my_invest.add_asset("META", 20, 160)
print(my_invest.holdings)
print(my_invest.total_value())
print(my_invest.most_valuable())
print(my_invest)
````

````python
import numpy as np

# Rows: years (2022, 2023, 2024), Columns: quarters (Q1, Q2, Q3, Q4)
revenue = np.array([
    [210, 230, 190, 250],
    [220, 260, 200, 270],
    [250, 290, 230, 310]])

# a)
print("a)")
years = [2022, 2023, 2024]
year_revenue = np.sum(revenue, axis = 1) #sum revenue by row, accroding to each year in the list

for year, rev in zip(years, year_revenue):
    print(f"revenue in {year}: {rev}")

# b)
print()
print("b)")
mean_qurt_rev = np.sum(revenue, axis = 1)/revenue.shape[1] #revenue.shape[1] find the number of columns
for year, mean in zip(years, mean_qurt_rev):
    print(f"mean_qurt_rev in {year}: {mean}")

# c)
print()
print("c)")

rev_2024 = revenue[2] 

growth_rate = np.array([1.05, 1.03, 1.08, 1.02])

print(f"adjusted values for 2024: {rev_2024*growth_rate}")

# d)
print()
print("d)")

years = [2022, 2023, 2024]
quarters = ["Q1", "Q2", "Q3", "Q4"]

print(np.where(revenue > 250))

rows, cols = np.where(revenue > 250) #create a boolean array satisfies revenue larger than 250

#loop and filter out which quarter in which year has revenue larger than 250
for r, c in zip(rows, cols):
    print(f"{quarters[c]} {years[r]}: {revenue[r, c]}")
````
